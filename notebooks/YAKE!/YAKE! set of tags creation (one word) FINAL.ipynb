## Function that implements YAKE!

```python
def get_keywords(text):
    
    language = "en"
    max_ngram_size = 1
    deduplication_thresold = 0.9
    deduplication_algo = 'seqm'
    windowSize = 1
    numOfKeywords = 20

    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(text)
    
    return keywords
```

## Get keywords, create 2 files: file with ticket id and it's tags and file that has a set of teh most common tags

```python
def add_to_ticket_and_tags(ticket_id, tag, value):
    min_value = 0.06
    max_value = 0.3
    
    if value <= min_value:
        return True
    elif value > min_value and value <= max_value:
        if any(s in ticket_id for s in data_human.keys()):
        #if ticket_id in data_human.keys():
            if tag in data_human[ticket_id]:
                return True
    return False
```

```python
tags_dict = {}
tickets_with_tags = {}
final_tags = []

for i in range(20000):
    text = data[i]['title'] + ' ' + data[i]['content']
    ticket_id = data[i]['id']
    
    # get tags generated by YAKE
    try:
        tags_list = get_keywords(text)
    except:
        print('An error occurred.')
        continue
    
    tags_for_ticket = []
    for tag_value_pair in tags_list:
        tag = tag_value_pair[0]
        value = tag_value_pair[1]
        
        if add_to_ticket_and_tags(ticket_id, tag, value):
            
            tags_for_ticket.append(tag)
            
            if tag in tags_dict:
                tags_dict[tag] += 1
            else:
                tags_dict[tag] = 1
                final_tags.append(tag)
                
    if ticket_id in data_human.keys():
        #find if there are any duplicate tags by human annotators
        unique_tags = []
        for tag in data_human[ticket_id]:
            if tag not in unique_tags:
                unique_tags.append(tag)
            else:
                if tag in tags_dict:
                    tags_dict[tag] += 1
                else:
                    tags_dict[tag] = 1
                    final_tags.append(tag)
                    
        
    tickets_with_tags[data[i]['id']] = tags_for_ticket
#print(tickets_with_tags)
#print(tags_dict)
```
